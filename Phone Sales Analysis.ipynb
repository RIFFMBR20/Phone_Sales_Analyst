```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phone_df = pd.read_csv('Sales.csv')
```

```python
phone_df.head()
```

```python
phone_df.tail()
```

```python
phone_df.isna().sum()
```

```python
phone_df = phone_df.dropna(subset=['Memory', 'Storage', 'Rating'])
```

```python
phone_df
```

```python
phone_df.describe()
```

```python
numerical_vars = ['Memory', 'Storage', 'Rating', 'Selling Price', 'Original Price']
for var in numerical_vars:
    plt.figure(figsize=(20, 10))
    sns.histplot(phone_df[var], kde=True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=phone_df, x='Brands')
plt.title('Count of Brands')
plt.xlabel('Brands')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
phone_df[phone_df['Brands'].isin(['SAMSUNG'])].value_counts().head(5)
```

```python
samsung_df = phone_df[phone_df['Brands'] == 'SAMSUNG']

top_mobiles = samsung_df['Mobile'].value_counts().head(10)

plt.figure(figsize=(10, 10))
sns.barplot(x=top_mobiles.index, y=top_mobiles.values)
plt.title('Top 10 Mobile Phones for Samsung')
plt.xlabel('Mobile')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
samsung_df = phone_df[phone_df['Brands'] == 'Apple']

top_mobiles = samsung_df['Mobile'].value_counts().head(10)

plt.figure(figsize=(10, 10))
sns.barplot(x=top_mobiles.index, y=top_mobiles.values)
plt.title('Top 10 Mobile Phones for Apple')
plt.xlabel('Mobile')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
samsung_df = phone_df[phone_df['Brands'] == 'realme']

top_mobiles = samsung_df['Mobile'].value_counts().head(10)

plt.figure(figsize=(10, 10))
sns.barplot(x=top_mobiles.index, y=top_mobiles.values)
plt.title('Top 10 Mobile Phones for realme')
plt.xlabel('Mobile')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
correlation_matrix = phone_df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
